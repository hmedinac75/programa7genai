Ejemplo Usando GPT 4

```python
from llama_index.llms.openai import OpenAI
from llama_index.core.chat_engine import SimpleChatEngine
```

```python
import os
with open("/content/apikey.txt") as archivo:
  apikey_openai = archivo.read()
```

```python
llm = OpenAI(temperature=0, model="gpt-4",api_key=apikey_openai)
chat_engine = SimpleChatEngine.from_defaults(llm=llm)
```

```python
response = chat_engine.chat("cuantos años tiene el mundo?")
response.response
```

Ejemplo unificado

```python
models = [
    {"class": OpenAI, "model": "gpt-4", "api_key": apikey_openai},
    {"class": Anthropic, "model": "claude-3-opus-20240229", "api_key": apikey_claude},
    {"class": Gemini, "model": "models/gemini-1.5-flash", "api_key": apikey_gemini},
]

# Pregunta que deseas enviar
question = "¿Cuántos años tiene el mundo?"

# Iterar sobre los modelos
for model_config in models:
    llm = model_config["class"](temperature=0, model=model_config["model"], api_key=model_config["api_key"])
    chat_engine = SimpleChatEngine.from_defaults(llm=llm)
    response = chat_engine.chat(question)
    print("--------")
    print(f"Modelo: {model_config['model']}")
    print(response.response)  # Muestra la respuesta de cada modelo
```
